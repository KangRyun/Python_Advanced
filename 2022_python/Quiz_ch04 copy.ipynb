각각의 Random walker는 주사위를 총 500번씩 던집니다.

```python
import random

num_walkers = 100
num_steps = 500

# 각 Walker의 초기 위치는 0으로 설정
walkers = [0] * num_walkers

for _ in range(num_steps):
    for i in range(num_walkers):
        # 주사위 던지기
        dice_roll = random.randint(1, 6)
        
        # 주사위 결과에 따라 위치 업데이트
        if dice_roll <= 3:
            walkers[i] -= 1  # 1, 2, 3이 나오면 뒤로 한칸
        else:
            walkers[i] += 1  # 4, 5, 6이 나오면 앞으로 한칸

walkers
# 결과 출력
#for i in range(num_walkers):
#    print(f"Walker {i + 1}: {walkers[i]}")
```

```python
import numpy as np

num_walkers = 100
num_steps = 500

# 주사위를 500번 던져 나오는 결과를 한 번에 생성합니다.
dice_rolls = np.random.randint(1, 7, size=(num_walkers, num_steps))

# 주사위 결과에 따라 이동 방향을 결정하고 누적 이동 거리를 계산합니다.
positions = np.where(dice_rolls <= 3, -1, 1)
cumulative_positions = np.cumsum(positions, axis=1)

# 각 Random walker의 최종 위치를 출력합니다.
#cumulative_positions
for i in range(num_walkers):
    final_position = cumulative_positions[i, -1]
    print(f"Walker {i+1}: Final position = {final_position}")
```

```python
import numpy as np

nwalkers = 100  # 랜덤 워커의 수
nsteps = 500  # 주사위를 던지는 횟수


# 랜덤 워커의 위치를 저장할 배열 생성
walkers = np.zeros((nwalkers, nsteps+1), dtype=int)

for i in range(nsteps):
    # 주사위 던지기
    dice = np.random.randint(1,7, size=nwalkers)
    
    # 1, 2, 3이 나오면 뒤로 한 칸 가기 (음수 방향)
    walkers[:, i+1] = walkers[:, i] - np.where(dice <= 3, 1, 0)
    
    # 4, 5, 6이 나오면 앞으로 한 칸 가기 (양수 방향)
    walkers[:, i+1] = walkers[:, i] + np.where(dice >= 4, 1, 0)

#결과 출력
for j in range(nwalkers):
    print(f"Walker {j+1}: {walkers[j]}")
```
