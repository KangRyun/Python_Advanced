### 4. 연속형 변수 이산형 변수로 변환하기

#### 나이(`Age`)를 10세 이하(young), 10초과-60세 미만(adult), 60세 초과-100세 이하(senior) 세 그룹으로 구분하여 Discretize(이산화) 시켜봅시다. 해당 데이터를 `Age_group` 이란 이름의 새로운 컬럼으로 저장하여 봅시다. 마지막으로, 그룹의 생존율을 계산해 봅시다.  
(5점)

Survived 값이 0이 사망한 사람이며, 1이 생존한 사람입니다.

```python
df['Age'].iloc[1] > 10
```

```python
type(df['Age'])
```

```python
Age_groups = []
for age in df['Age']:
    if age <= 10:
        Age_group = 'Young'
    elif age < 60:
        Age_group = 'Adult'
    else:
        Age_group = 'Senior'
    Age_groups.append(Age_group)
```

```python
df['Age_groups'] = Age_groups
```

```python
S = df[df.Age_groups == 'Senior']
A = df[df.Age_groups == 'Adult']
Y = df[df.Age_groups == 'Young']
```

```python
(len(S[S.Survived == 1]) / len(S[['Survived']])) * 100
```

```python
(len(A[A.Survived == 1]) / len(A[['Survived']])) * 100
```

```python
(len(Y[Y.Survived == 1]) / len(Y[['Survived']])) * 100
```
