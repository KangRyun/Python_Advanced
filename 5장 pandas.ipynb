### Function Application and Mapping

```python
frame = pd.DataFrame(np.random.standard_normal((4, 3)),
                     columns=list("bde"),
                     index=["Utah", "Oiho", "Texas", "Oregon"])
frame
```

```python
np.abs(frame)
```

```python
def f1(x):
    return x.max() - x.min()
```

```python
frame.apply(f1)
```

```python
frame.max() - frame.min()
```

```python
frame.max()
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
frame.apply(f2)
```

```python
def my_format(x):
    return f"{x:.2f}" # 소수점 둘째자리까지 포맷팅된 문자열을 반환합니다.
```

```python
frame.applymap(my_format) # frame의 모든 요소에 `my_format` 함수를 적용합니다.
```

```python
frame["e"].map(my_format)
```

### Correlation and Covariance

```python
price = pd.read_pickle("yahoo_price.pkl")
volume = pd.read_pickle("yahoo_volume.pkl")
```

```python
returns = price.pct_change()
returns.tail()
```

```python
returns["MSFT"].corr(returns["IBM"]) # 상관관계
```

```python
returns["MSFT"].cov(returns["IBM"]) # 공분산
```

```python
returns.corr()
```

```python
returns.corrwith(returns["IBM"])
```

### Unique Values, Value Counts, and Membership

```python
obj = pd.Series(["c", "a", "d", "a", "a", "b", "b", "c", "c"])
uniques = obj.unique()
uniques # Series에서 중복을 제외한 고유한 값들을 추출하여 배열로 반환
```

```python
obj.value_counts() # Series에서 각 값들의 개수를 세어 반환
```

```python
pd.value_counts(obj.to_numpy(), sort=False) # Series를 NumPy 배열로 변환한 뒤 각 값들의 개수를 세어 반환
```
