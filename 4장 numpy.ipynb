```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
result = []
for x, y, c in zip(xarr, yarr, cond):
    if c:
        result.append(x)
    else:
        result.append(y)
result
```

```python
result = [(x if c else y)
          for x, y, c in zip(xarr, yarr, cond)]
result
```

```python
import random
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
nsteps = 1000
rng = np.random.default_rng(seed=12345)
draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws == 0, 1, -1)
walk = steps.cumsum()
```

```python
nwalks = 5000 # 랜덤 워크 시뮬레이션을 수행할 워크(걷기)의 수입니다. 이 예시에서는 5000개의 워크가 생성됩니다.
nsteps = 1000 # 한 워크에서 수행할 총 단계 수입니다. 각 워크는 이 총 단계 수에 따라 이동합니다. 이 예시에서는 1000단계로 설정되어 있습니다.
draws = rng.integers(0, 2, size=(nwalks, nsteps)) # rng.integers(0, 2, size=(nwalks, nsteps))를 통해 0과 1 사이의 정수를 무작위로 생성한 배열입니다. 
                                                    # 이 배열은 nwalks 행과 nsteps 열을 가지며, 각 워크에서 각 단계마다 무작위로 0 또는 1이 선택됩니다.
steps = np.where(draws > 0, 1, -1) # np.where(draws > 0, 1, -1)를 사용하여 draws 배열의 각 요소가 0보다 큰 경우 1을, 그렇지 않은 경우 -1을 가지는 배열을 생성합니다. 
                                    # 이 배열은 draws 배열과 같은 크기를 가지며, 각 단계에서의 이동 방향을 나타냅니다. 
                                    # 0보다 큰 경우에는 1이 선택되어 오른쪽으로 이동하고, 0인 경우에는 -1이 선택되어 왼쪽으로 이동합니다.
walks = steps.cumsum(axis=1) # steps.cumsum(axis=1)를 통해 steps 배열을 누적 합(cumulative sum)하여 워크의 위치를 나타내는 배열을 생성합니다. 
                                # axis=1은 각 행에 대해 누적 합을 계산함을 의미합니다. 
                                # 이로써 walks 배열은 nwalks 행과 nsteps 열을 가지며, 각 워크의 각 단계에서의 위치를 나타냅니다.
```

```python
plt.plot(walks[0], label='Walk 1')
plt.plot(walks[1], label='Walk 2')
plt.plot(walks[2], label='Walk 3')
plt.plot(walks[3], label='Walk 4')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.legend()
plt.show()
```
