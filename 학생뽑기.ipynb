#### Quiz 1 - 결석한 학생들은 빼고 학생을 출력되도록 업데이트 해봅시다. 

```python
absence=['조성택','최원준','정대규','이강륜','노건우'
         ,'김재현','김희수','최송연']
```

```python
import numpy as np
import pandas as pd

# 예시 데이터프레임을 생성
data = pd.read_csv('pandas_practice.csv')
df = pd.DataFrame(data)

# 결석한 학생 리스트를 저장
absent_students = ['조성택','최원준','정대규','이강륜','노건우'
         ,'김재현','김희수','최송연']

def pop_student(df, absent_students):
    # 결석한 학생들을 제외한 학생들만 선택
    present_students = df[~df['이름'].isin(absent_students)]

    # 무작위로 학생 선택
    length = present_students.shape[0]
    rand_num = np.random.randint(0, length)
    name = present_students.iloc[rand_num]["이름"]

    print(f"       축하합니다! \n{name} 학생이 당첨 되었습니다.")

# 함수
pop_student(df, absent_students)
```
